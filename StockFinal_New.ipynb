```python
#分開train and test
df_Train = dfTotal[~dfTotal['0056_close_y'].isna()]
df_Test  = dfTotal[dfTotal['0056_close_y'].isna()]
```

```python
df_Test.isna().sum()
#df_Train.isna().sum()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.svm import SVR
```

```python
X = df_Train.drop('0056_close_y', axis = 1)
Y = df_Train['0056_close_y']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.15, random_state = 5)
```

```python
modelSvr = SVR(kernel='rbf', C=1000)
modelSvr.fit(X_train, Y_train)
pred_Svr = modelSvr.predict(X_test)

plt.scatter(Y_test, pred_Svr)
plt.plot([200, 400], [200, 400], 'r')
plt.title(f'mse of svm = {mean_squared_error(Y_test, pred_Svr)}, r2={r2_score(Y_test, pred_Svr)}')
```

```python
# test_prediction 就是預測結果
test_set = df_Test.drop('0056_close_y', axis = 1)
test_prediction = modelSvr.predict(test_set) 
```

```python
test_prediction = pd.DataFrame(test_prediction)
test_prediction.columns = ["prediction"]
test_prediction.index = test_set.index
output=pd.DataFrame(data={"date":test_set.index,"0056_close_y":test_prediction["prediction"]})
output.to_csv(path_or_buf="my_submission.csv",index=False,quoting=3,sep=',')
output
```
